```python
#extracting the data
X=df[['Name','Location','Year','Kilometers_Driven','Fuel_Type','Transmission','Owner_Type','Mileage','Engine','Seats']]
y=df['Price']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
ohe=OneHotEncoder()
ohe.fit(X[['Name','Location','Fuel_Type','Transmission','Owner_Type']])
```

```python
column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['Name','Location','Fuel_Type','Transmission','Owner_Type']),
                                    remainder='passthrough')
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(column_trans,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
from sklearn import metrics
import numpy as np
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
```

```python
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
